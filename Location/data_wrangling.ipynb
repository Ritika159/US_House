```python
df_unpivot = pd.melt(df,id_vars=['Rank','Region Name','State Name','Latitude','Longitude'],value_vars=['2000', '2001', '2002', '2003',
       '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012',
       '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020', '2021',
       '2022'],var_name='Year',value_name='Average House Price')
```
