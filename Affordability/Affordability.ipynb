```python
home_price_affordability = pd.read_csv("G:\My Drive\Data Science and Machine Learning\ML and DS Projects\Projects\Hoom\Affordability\Home Price Affordability.csv")
home_price_affordability.head()
```

```python
fig = make_subplots(specs=[[{"secondary_y": True}]])

#Add traces
fig.add_trace(go.Scatter(x=home_price_affordability['Date'],y=home_price_affordability['Home Price Affordability Median HHI'],name='Home Price Affordability Median HHI'),secondary_y=False)
fig.add_trace(go.Scatter(x=home_price_affordability['Date'],y=home_price_affordability['Home Price Affordability 25% HHI'],name='Home Price Affordability 25% HHI'),secondary_y=False)
fig.add_trace(go.Scatter(x=home_price_affordability['Date'],y=home_price_affordability['Home Price Affordability 75% HHI'],name='Home Price Affordability 75% HHI'),secondary_y=False)
fig.add_trace(go.Scatter(x=home_price_affordability['Date'],y=home_price_affordability['Median Home Price'],name='Median Home Price'),secondary_y=True)

#Title
fig.update_layout(title_text='Home Price Affordability in the United States',height=500)

#X-axis title
fig.update_xaxes(title_text='Date')

#Y-axis titles
fig.update_yaxes(title_text='Home Affordability',secondary_y=False)
fig.update_yaxes(title_text='Median Home Price',secondary_y=True)

fig.show()
```
